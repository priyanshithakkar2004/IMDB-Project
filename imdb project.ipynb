##  REMOVE DUPLICATES


```python
#  Remove duplicate rows to clean the dataset
df = df.drop_duplicates().copy()
```

```python
# Drop columns that are not useful for analysis
columns_to_remove = ['homepage', 'backdrop_path', 'poster_path', 'imdb_id', 'tagline']
df = df.drop(columns=columns_to_remove)
```

```python
#remove duplicates of title and release_date, as they likely represent the same movie
df = df.drop_duplicates(subset=['title', 'release_date'])
df

```

## Fixing Data Types

```python
# Convert 'release_date' to proper datetime format
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
#Extract year from release_date for analysis
df['YEAR'] = df['release_date'].dt.year
```

```python
#Convert important columns to numeric values
numeric_cols = ['vote_average', 'vote_count', 'runtime', 'budget', 'revenue', 'popularity']

for col in numeric_cols:
    df.loc[:, col] = pd.to_numeric(df[col], errors='coerce')
```

## Handling Missing Values

```python
 # Fill missing values in numeric columns with meaningful defaults
df = df.fillna({
    'vote_average': df['vote_average'].mean(),   # use average rating
    'vote_count': 0,                             # no votes → 0
    'runtime': df['runtime'].median(),           # typical movie length
    'budget': 0,
    'revenue': 0,
    'popularity': df['popularity'].median()
})
```

```python
#Remove rows where YEAR is missing (important for time-based analysis)
df = df.loc[df['YEAR'].notna()].copy()
```

```python
# Fill missing values in text columns
df['title'] = df['title'].fillna("Unknown")
df['original_title'] = df['original_title'].fillna("Unknown")
df['overview'] = df['overview'].fillna("No overview")
df['genres'] = df['genres'].fillna("Unknown")
```

```python
# Extract main genre (first genre from the list)
df['main_genre'] = df['genres'].apply(lambda x: str(x).split(',')[0])
```

```python
df
```

## Feature Engineering

```python
# Calculate profit (revenue - budget)
df['profit'] = df['revenue'] - df['budget']
```

```python
# Calculate ROI (Return on Investment)
# Avoid division by zero by checking budget
df['ROI'] = df.apply(
    lambda row: row['revenue'] / row['budget'] if row['budget'] > 0 else 0,axis=1)
```

```python
#Create decade column (e.g., 1995 → 1990s)
df['decade'] = (df['YEAR'] // 10) * 10
```

```python
#Extract main genre (first genre from the list)
df['main_genre'] = df['genres'].astype(str).apply(lambda x: x.split(',')[0])
```

## EDA Part A: Revenue & Budget Analysis

```python
# ── Plot 1: Budget vs Revenue Scatter Plot ──

plt.figure(figsize=(8, 5))

plt.scatter(df['budget'] / 1e6, df['revenue'] / 1e6,
            alpha=0.4, color='steelblue', s=15,
            edgecolors='white', linewidths=0.3)

plt.xlabel('Budget ($ Millions)', fontsize=12)
plt.ylabel('Revenue ($ Millions)', fontsize=12)
plt.title('Budget vs Revenue', fontsize=14, fontweight='bold')

plt.grid(True, linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# ── Plot 2: Profit Distribution ──

plt.figure(figsize=(8, 5))

plt.hist(df['profit'] / 1e6, bins=50, color='green', edgecolor='white')

plt.axvline(df['profit'].mean() / 1e6, color='red', linestyle='--', linewidth=1.5, label='Mean Profit')

plt.xlabel('Profit ($ Millions)', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.title('Profit Distribution', fontsize=14, fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# ── Plot 3: ROI Distribution (with outliers removed) ──
q_low = df['ROI'].quantile(0.01)
q_high = df['ROI'].quantile(0.99)
filtered_df = df[(df['ROI'] > q_low) & (df['ROI'] < q_high)]
plt.figure(figsize=(10, 5))
plt.hist(filtered_df['ROI'], bins=50)
plt.title('ROI Distribution', fontsize=14, fontweight='bold')
plt.xlabel('ROI', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# ── Plot 4: Correlation Heatmap ──
cols = ['budget', 'revenue', 'vote_average', 'profit', 'ROI', 'popularity', 'runtime']
corr = df[cols].corr()

plt.figure(figsize=(10, 6))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.tight_layout()
plt.show()
```

```python
# ── Plot 5: Top 10 Most Profitable Movies ──

top10 = df.nlargest(10, 'profit')[['title', 'profit']]

plt.figure(figsize=(15, 5))
plt.barh(top10['title'], top10['profit'] / 1e6, color='steelblue')
plt.xlabel('Profit ($ Millions)')
plt.title('Top 10 Most Profitable Movies')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

## EDA Part B: Genre Analysis

```python
# Split genres (e.g. 'Action|Drama' → separate rows)
df_genre = df.copy()
df_genre['genres'] = df_genre['genres'].str.split('|')
df_genre = df_genre.explode('genres')
df_genre['genres'] = df_genre['genres'].str.strip()

print('Genre dataset created!')
print('Unique genres:', df_genre['genres'].nunique())
```

```python
# ── Plot 6: Genre Frequency ──

genre_count = df_genre['genres'].value_counts().head(10)

plt.figure(figsize=(10, 5))
genre_count.plot(kind='bar', color='coral', edgecolor='white')

plt.xlabel('Genre', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.title('Top 10 Genres by Movie Count', fontsize=14, fontweight='bold')

plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
# ── Plot 7: Genre vs Average Revenue  ──
genre_revenue = df_genre.groupby('genres')['revenue'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))

plt.bar(genre_revenue.index, genre_revenue.values / 1e6,
        color='teal')

plt.xlabel('Genre', fontsize=12)
plt.ylabel('Average Revenue ($ Millions)', fontsize=12)
plt.title('Top 10 Genres by Average Revenue', fontsize=14, fontweight='bold')

plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

```python
# ── Plot 8: Genre vs Average Profit ──

# First check if df_genre exists and has data
df_genre = df.copy()
df_genre['genres'] = df_genre['genres'].fillna('Unclassified')
df_genre['genres'] = df_genre['genres'].str.split(', ')
df_genre = df_genre.explode('genres')
df_genre['genres'] = df_genre['genres'].str.strip()

print('df_genre rows:', len(df_genre))
print('Sample genres:', df_genre['genres'].unique()[:5])

# Now plot
genre_profit = df_genre.groupby('genres')['profit'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 5))
genre_profit.div(1e6).plot(kind='barh', color='green', edgecolor='white')
plt.xlabel('Average Profit ($ Millions)')
plt.title('Average Profit by Genre')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```

```python
# ── Plot 9: Genre Trends Over Decades ──

df['decade'] = (df['YEAR'] // 10) * 10

df_filtered = df[(df['decade'] >= 1900) & (df['decade'] <= 2020)]

top_genres = df_filtered['main_genre'].value_counts().head(5).index
filtered_df = df_filtered[df_filtered['main_genre'].isin(top_genres)]

genre_trend = filtered_df.groupby(['decade', 'main_genre']).size().reset_index(name='count')

pivot_df = genre_trend.pivot(index='decade', columns='main_genre', values='count').fillna(0)

pivot_df.plot(kind='bar', figsize=(12, 6))

plt.title('Genre Trends Over Decades (1900–2020)', fontsize=14, fontweight='bold')
plt.xlabel('Decade', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)

plt.xticks(rotation=45)
plt.legend(title='Genre', fontsize=9)

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

## EDA Part C: Time Series Analysis

```python
print(df.columns)
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
df['release_year'] = df['release_date'].dt.year
```

```python
# Time Series Analysis

df_time = df[df['YEAR'].between(1980, 2024)]

yearly = df_time.groupby('YEAR').agg(
    movie_count    = ('title', 'count'),
    total_revenue  = ('revenue', 'sum'),
    avg_rating     = ('vote_average', 'mean'),
    avg_popularity = ('popularity', 'mean')
).reset_index()

(yearly.head())
```

```python
# ── Plot 10: Movies Released per Year ──

plt.figure(figsize=(12, 4))

plt.plot(yearly['YEAR'], yearly['movie_count'], color='steelblue', linewidth=2, marker='o')
plt.fill_between(yearly['YEAR'], yearly['movie_count'], alpha=0.3, color='steelblue')

plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Movies', fontsize=12)
plt.title('Movies Released Per Year', fontsize=14, fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python

# ── Plot 11: Revenue Trend Over Years ──
yearly = yearly.rename(columns={'YEAR': 'year'})

plt.figure(figsize=(12, 4))

plt.plot(yearly['year'], yearly['total_revenue'] / 1e9, color='green', linewidth=2, marker='o')

plt.xlabel('Year', fontsize=12)
plt.ylabel('Total Revenue ($ Billions)', fontsize=12)
plt.title('Total Revenue Trend Over Years', fontsize=14, fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.savefig('plot11_revenue_trend.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# ── Plot 12: Average Rating Trend ──

plt.figure(figsize=(12, 4))

plt.plot(yearly['year'], yearly['avg_rating'], color='orange', linewidth=2, marker='o', markersize=3)

plt.axhline(yearly['avg_rating'].mean(), color='red', linestyle='--', linewidth=1.5, label='Overall Average')

plt.xlabel('Year', fontsize=12)
plt.ylabel('Avg Rating', fontsize=12)
plt.title('Average Rating Trend Over Years', fontsize=14, fontweight='bold')

plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()

plt.tight_layout()
plt.show()
```
